## Count ROH

```python
%%R

# Define the ancestries to loop through
ancestries <- c("AAC", "AFR", "AJ", "AMR", "CAS", "EAS", "EUR", "MDE", "SAS")

# Initialize an empty dataframe to store the summary statistics
summary <- data.frame()

# Loop through each ancestry to calculate the summary statistics
for(ANC in ancestries) {
    # Read in the ROH data for the current ancestry
    data <- read.table(paste0(ANC, "_ROH.hom"), header = TRUE)
    
    # Rename the 10th column to 'SNPs' for clarity
    colnames(data)[10] <- "SNPs"
    
    # Calculate the summary statistics and append to the summary dataframe
    row <- data %>%
        summarise(ANC = ANC,
                  Total = n(), 
                  minKB = min(KB), 
                  maxKB = max(KB),  
                  meanKB = mean(KB),  
                  minSNP = min(SNPs), 
                  maxSNP = max(SNPs), 
                  meanSNP = mean(SNPs))
    
    # Bind the result to the summary dataframe
    summary <- rbind(summary, row)
}

# Display the summary dataframe
summary
```

```python
%%R

# Summarize all the ancestries by calculating overall summary statistics
ALL <- summary %>%
    summarise(
        ANC = "ALL",
        Total = sum(Total), 
        minKB = min(minKB), 
        maxKB = max(maxKB),  
        meanKB = mean(meanKB),
        minSNP = min(minSNP), 
        maxSNP = max(maxSNP),  
        meanSNP = mean(meanSNP)
    )

# Display the combined summary for all ancestries
ALL
```

## Percentage with ROH

```python
%%R

# Define the ancestries to loop through
ancestries <- c("AAC", "AFR", "AJ", "AMR", "CAS", "EAS", "EUR", "MDE", "SAS")

# Initialize an empty dataframe for summary statistics
summary <- data.frame(NULL)

# Loop through each ancestry
for (ANC in ancestries) {
  
  # Read in the .hom.indiv file for each ancestry
  data <- read.table(paste0(ANC, "_ROH.hom.indiv"), header = TRUE)
  
  # Filter out individuals with PHE value of -9 (usually missing phenotype)
  data <- data[data$PHE != -9, ]
  
  # Create a new column: 1 if the individual harbours ROH (NSEG > 0), else 0
  data$harbourROH <- ifelse(data$NSEG == 0, 0, 1)
  
  # Summarise the data by PHE (cases/controls), counting individuals who harbour ROH
  row <- data %>%
    group_by(PHE) %>%
    summarise(
      ANC = ANC,  # Include the ancestry
      Total = n(),  # Total number of individuals
      harbourROH = sum(harbourROH),  # Count of individuals with ROH
      Percentage = harbourROH / Total  # Percentage of individuals with ROH
    )
  
  # Append the summarised row to the summary dataframe
  summary <- rbind(summary, row)
}

# Display the final summary
summary
```

```python
%%R

# Pivot the summary dataframe to a wide format, separating controls and cases
data_wide <- summary %>%
  pivot_wider(names_from = PHE, values_from = c(Total, harbourROH, Percentage), names_prefix = "PHE") %>%
  rename(
    Controls = Total_PHE1,             # Rename Total_PHE1 to Controls
    Cases = Total_PHE2,                # Rename Total_PHE2 to Cases
    Controls.ROH = harbourROH_PHE1,    # Rename harbourROH_PHE1 to Controls.ROH
    Cases.ROH = harbourROH_PHE2,       # Rename harbourROH_PHE2 to Cases.ROH
    Controls.Percentage = Percentage_PHE1,  # Rename Percentage_PHE1 to Controls.Percentage
    Cases.Percentage = Percentage_PHE2      # Rename Percentage_PHE2 to Cases.Percentage
  )
data_wide

# Summarise totals across all ancestries (ALL row)
all_row <- data.frame(
  ANC = "ALL",  # All ancestries combined
  Controls = as.numeric(sum(data_wide$Controls)),       # Total controls
  Cases = as.numeric(sum(data_wide$Cases)),             # Total cases
  Controls.ROH = sum(data_wide$Controls.ROH),           # Total controls with ROH
  Cases.ROH = sum(data_wide$Cases.ROH)                  # Total cases with ROH
)

# Calculate percentages for the combined row
all_row$Controls.Percentage <- all_row$Controls.ROH / all_row$Controls
all_row$Cases.Percentage <- all_row$Cases.ROH / all_row$Cases

# Combine the ALL row with the rest of the data
all_data <- rbind(data_wide, all_row)

# Calculate the ratio of Cases to Controls percentages
all_data$ratio <- all_data$Cases.Percentage / all_data$Controls.Percentage

# Initialize a dataframe for storing p-values
pvalue_column <- data.frame(NULL)

# List of ancestries including the combined "ALL" group
ancestries <- c("AAC", "AFR", "AJ", "AMR", "CAS", "EAS", "EUR", "MDE", "SAS", "ALL")

# Perform a proportion test for each ancestry
for (ANC in 1:length(ancestries)) {
  cases_ROH <- all_data$Cases.ROH[ANC]
  control_ROH <- all_data$Controls.ROH[ANC]
  cases <- all_data$Cases[ANC]
  controls <- all_data$Controls[ANC]
  
  # Proportion test to check the significance of ROH differences between cases and controls
  pvalue <- prop.test(x = c(cases_ROH, control_ROH), n = c(cases, controls))$p.value
  pvalue_column <- rbind(pvalue_column, pvalue)  # Append p-values to the dataframe
}

# Add the p-values to the final all_data dataframe
all_data <- cbind(all_data, pvalue_column)
colnames(all_data)[9] <- "pvalue"  # Rename the new column to 'pvalue'

# Save the results to a file
fwrite(all_data, "percentage_ROH.tsv", sep = '\t')
```

## FROH in all ancestries

```python
%%R

# Define the ancestries to loop through
ancestries <- c("AAC", "AFR", "AJ", "AMR", "CAS", "EAS", "EUR", "MDE", "SAS")

# Initialize an empty data frame to store FROH results for all ancestries
data_all <- data.frame(NULL)

# Loop through each ancestry to calculate FROH
for (ANC in 1:length(ancestries)) {
  
  # Read the ROH data (.hom.indiv file) for the current ancestry
  data <- read.table(paste0(ancestries[ANC], "_ROH.hom.indiv"), header = TRUE)
  
  # Calculate FROH by dividing the total ROH length (KB) by the total autosomal genome length (2,886,428 Kb)
  data$FROH <- round(data$KB / 2886428, digits = 5)  # 5 digits precision
  
  # Summarize the mean FROH value for the current ancestry
  row <- data %>%
    summarise(ANC = ancestries[ANC], FROH = mean(FROH))  # Calculate mean FROH
  
  # Append the result to the data_all data frame
  data_all <- rbind(data_all, row)
}

# Write the final FROH data to a .tsv file
fwrite(data_all, "FROH_in_all_ancestries.tsv", sep = '\t')
```

```python
%%R

# Function to load inbreeding and phenotype data, merge, and assign ancestry
load_and_merge_data <- function(ancestry) {
  inbreeding_data <- fread(paste0(ancestry, "_inbreeding.het"))[, .(IID, F)]  # Load inbreeding data (IID, F)
  phenotype_data <- fread(paste0(ancestry, "_release6.psam"))[, .(`#IID`, PHENO1)]  # Load phenotype data
  setnames(phenotype_data, old = "#IID", new = "IID")  # Rename column to match inbreeding data
  setnames(phenotype_data, old = "PHENO1", new = "PHENO")
  
  # Merge inbreeding data and phenotype data
  merged_data <- merge(inbreeding_data, phenotype_data, by = "IID")
  
  # Add ancestry column
  merged_data[, Ancestry := ancestry]
  
  return(merged_data)
}

# List of ancestries
ancestries <- c("AJ", "AAC", "AFR", "AMR", "CAS", "EAS", "MDE", "SAS", "EUR")

# Apply the function to each ancestry and store the result in a list
ancestry_data_list <- lapply(ancestries, load_and_merge_data)

# Combine all ancestry data into one data.table
combined_data <- rbindlist(ancestry_data_list)
```

## Extract Age ( Mean ± SE) for Each Ancestry

```python
%%R

# List of ancestries to loop through
ancestries <- c("AAC", "AFR", "AJ", "AMR", "CAS", "EAS", "EUR", "MDE", "SAS")

# Initialize an empty data frame to store results
results <- data.frame()

# Loop over each ancestry
for (ANC in ancestries) {
    
    # Load individual ROH data for the specific ancestry
    data <- fread(paste0(ANC, "_ROH.hom.indiv"), header = TRUE)
    
    # Filter data to exclude rows with missing phenotype (-9)
    dat <- data[data$PHE != -9, ]
    
    # Recode phenotype: subtract 1 to convert PHE values from 1/2 to 0 (control) / 1 (case)
    dat$PHE <- dat$PHE - 1
    
    # Load COVAR data for the same ancestry
    cov <- fread(paste0(ANC, "_COVAR.txt"), header = TRUE)
    
    # Merge the ROH data with COVAR data by "IID"
    merged <- merge(dat, cov, by = "IID")
    
    # Impute missing age: replace NA values with the mean age (rounded to nearest integer)
    merged_impute_age <- merged
    merged_impute_age[is.na(merged$age), ]$age <- round(mean(merged$age, na.rm = TRUE), 0)
    
    # Calculate statistics:
    # - ALL: Total number of individuals with age data
    # - MISSING: Number of individuals with missing age before imputation
    # - MEAN: Mean age after imputing missing values
    # - SD: Standard deviation of age
    ALL <- length(merged$age)
    MISSING <- sum(is.na(merged$age))
    MEAN <- round(mean(merged$age, na.rm = TRUE), 0)
    SD <- round(sd(merged$age, na.rm = TRUE), 0)
    
    # Consolidate results for the current ancestry
    row <- data.frame(ANC = ANC, N = ALL, Missing_age = MISSING, Mean = MEAN, Sd = SD)
    results <- rbind(results, row)
}

# Set the column names for the results data frame
colnames(results) <- c("ANC", "N", "Missing_age", "Mean", "SD")

# Write the results of the missing age analysis to a file for each ancestry
fwrite(results, "missing_age_summary.txt")
```

## ROH Table - Cases vs. controls

### T-test: mean values (Sup. Table 7)

```python
%%R

# List of ancestries to loop through
ancestries <- c("AAC", "AFR", "AJ", "AMR", "CAS", "EAS", "EUR", "MDE", "SAS")

# Initialize an empty data frame to store results
results <- data.frame()

# Loop through each ancestry
for (ANC in ancestries) {
    
    # Load individual ROH data with FROH for the specific ancestry
    data <- fread(paste0(ANC, "_ROH_withFROH.hom.indiv"), header = TRUE)
    
    # Recode Cases and Controls: PHE == -9 is missing data, so we filter those out
    dat <- data[data$PHE != -9, ]
    dat$PHE <- dat$PHE - 1  # Recode phenotype: 1 -> control (0), 2 -> case (1)
    
    # Summarize data by phenotype (case/control) using group_by and summarise from dplyr
    data_tidy <- dat %>%
        group_by(PHE) %>%
        summarise(
            NSEG_mean = mean(NSEG, na.rm = TRUE),
            KB_mean = mean(KB, na.rm = TRUE),
            KBAVG_mean = mean(KBAVG, na.rm = TRUE),
            NSEG_se = se(NSEG, na.rm = TRUE),
            KB_se = se(KB, na.rm = TRUE),
            KBAVG_se = se(KBAVG, na.rm = TRUE),
            FROH_mean = mean(FROH, na.rm = TRUE),
            FROH_se = se(FROH, na.rm = TRUE)
        )
    
    # Perform t-tests for the four parameters
    model1_imp <- t.test(NSEG ~ PHE, data = dat)    # NROH
    model2_imp <- t.test(KB ~ PHE, data = dat)      # SROH
    model3_imp <- t.test(KBAVG ~ PHE, data = dat)   # AVROH
    model4_imp <- t.test(FROH ~ PHE, data = dat)    # FROH
    
    # Calculate mean, SE, and p-value for each parameter
    # SROH (Sum of ROH lengths)
    SROH_cases <- paste(round(data_tidy$KB_mean[2], 1), "±", round(data_tidy$KB_se[2], 1))
    SROH_controls <- paste(round(data_tidy$KB_mean[1], 1), "±", round(data_tidy$KB_se[1], 1))
    SROH_pvalue <- signif(model2_imp$p.value, 3)
    
    # AVROH (Average ROH length)
    AVROH_cases <- paste(round(data_tidy$KBAVG_mean[2], 1), "±", round(data_tidy$KBAVG_se[2], 1))
    AVROH_controls <- paste(round(data_tidy$KBAVG_mean[1], 1), "±", round(data_tidy$KBAVG_se[1], 1))
    AVROH_pvalue <- signif(model3_imp$p.value, 3)
    
    # NROH (Number of ROH segments)
    NROH_cases <- paste(round(data_tidy$NSEG_mean[2], 1), "±", round(data_tidy$NSEG_se[2], 1))
    NROH_controls <- paste(round(data_tidy$NSEG_mean[1], 1), "±", round(data_tidy$NSEG_se[1], 1))
    NROH_pvalue <- signif(model1_imp$p.value, 3)
    
    # FROH (Genomic inbreeding coefficient)
    FROH_cases <- paste(round(data_tidy$FROH_mean[2], 4), "±", signif(data_tidy$FROH_se[2], 1))
    FROH_controls <- paste(round(data_tidy$FROH_mean[1], 4), "±", signif(data_tidy$FROH_se[1], 1))
    FROH_pvalue <- signif(model4_imp$p.value, 3)
    
    # Consolidate results for each ancestry
    row <- c(ANC, SROH_cases, SROH_controls, SROH_pvalue, 
             AVROH_cases, AVROH_controls, AVROH_pvalue, 
             NROH_cases, NROH_controls, NROH_pvalue, 
             FROH_cases, FROH_controls, FROH_pvalue)
    
    # Append the row to the results data frame
    results <- rbind(results, row)
}

# Assign column names to the results data frame
colnames(results) <- c("Ancestry", "SROH_cases", "SROH_controls", "SROH_pvalue", 
                       "AVROH_cases", "AVROH_controls", "AVROH_pvalue", 
                       "NROH_cases", "NROH_controls", "NROH_pvalue", 
                       "FROH_cases", "FROH_controls", "FROH_pvalue")

# Save the results to a text file
fwrite(results, "Pvalues_T-test_casesVScontrols.txt")

# Display the results data frame
results
```

```python
%%R

library(data.table) # For fread() and fwrite()
library(stringr)    # For str_split()

# Define ancestries
ancestries <- c("AAC", "AFR", "AJ", "AMR", "CAS", "EAS", "EUR", "MDE", "SAS")

# Load ancestry cases/controls data
ancestry_cases_controls <- fread("ancestry_cases_controls.txt", header = TRUE)

# Loop through each ancestry
for (ANC in 1:length(ancestries)) {
  
  # Load the ROH data for the current ancestry
  ROH <- fread(paste0(ancestries[ANC], "_ROH.hom.overlap.CON.PDgenes.txt"), header = TRUE)
  
  # Split 'case_to_control' into cases and controls
  ROH[, c("case", "control") := tstrsplit(case_to_control, ":", type.convert = TRUE)]
  
  # Convert cases and controls to numeric
  ROH[, `:=`(NcasesWithROHs = as.numeric(case), 
             NcontrolsWithROHs = as.numeric(control))]
  
  # Get total cases and controls for the ancestry
  total_cases <- ancestry_cases_controls[Ancestry == ancestries[ANC], Cases]
  total_controls <- ancestry_cases_controls[Ancestry == ancestries[ANC], Controls]
  
  # Add additional columns (in separate steps to avoid referencing undefined columns)
  ROH[, caseN := NcasesWithROHs]
  ROH[, controlN := NcontrolsWithROHs]
  ROH[, combinedN := caseN + controlN]
  ROH[, propCases := caseN / total_cases]
  ROH[, propControls := controlN / total_controls]
  
  # Initialize a p-value column
  ROH[, P := NA_real_]
  
  # Perform proportion tests for each row
  for (i in 1:nrow(ROH)) {
    test_result <- prop.test(
      x = c(ROH$NcasesWithROHs[i], ROH$NcontrolsWithROHs[i]), 
      n = c(total_cases, total_controls)
    )
    ROH$P[i] <- test_result$p.value
  }
  
  # Identify case-enriched ROHs
  ROH[, caseEnriched := ifelse(propCases > propControls, 1, 0)]
  
  # Save the full ROH data
  fwrite(ROH, paste0(ancestries[ANC], "_ROH_all.txt"))
  
  # Subset to case-enriched ROHs with a total count >= 1
  ROH_subsetted <- ROH[combinedN >= 1 & caseEnriched == 1]
  
  # Check for zero genes before applying Bonferroni correction
  Ngenes <- nrow(ROH_subsetted)
  if (Ngenes > 0) {
    ROH_subsetted[, passMultiTest := ifelse(P <= (0.05 / Ngenes), 1, 0)]
  } else {
    ROH_subsetted[, passMultiTest := 0]
  }
  
  # Save the subset of case-enriched ROHs
  fwrite(ROH_subsetted, paste0(ancestries[ANC], "_ROH_caseEnriched.txt"))
}
```

```python
%%R
library(data.table)

# Set working directory
setwd("/home/jupyter/ROH_plots/")

# Define the ancestries and ROH ranges
ancestries <- c("AAC", "AFR", "AJ", "AMR", "CAS", "EAS", "EUR", "MDE", "SAS")
roh_ranges <- c("2kb", "3kb", "4kb", "5kb", "6kb", "7kb", "8kb", "9kb", "10kb")

# Loop through each ancestry and ROH range
for (ancestry in ancestries) {
  for (roh_range in roh_ranges) {
    # Define file paths
    roh_file <- sprintf("%s_ROH_%s.hom.indiv", ancestry, roh_range)  # Fixed format specifier
    clinical_file <- "master_key_release6_final.csv"
    pca_file <- sprintf("%s_release6.eigenvec", ancestry)

    # Read clinical and PCA data
    clinical <- fread(clinical_file)
    colnames(clinical)[2] <- "IID"
    pca <- fread(pca_file)

    # Merge clinical and PCA data
    merged <- merge(clinical, pca, by = "IID")
    COVAR <- merged[, c("IID", "sex_for_qc", "phenotype_for_qc", "age", "age_of_onset", "PC1", "PC2", "PC3",
                        "PC4", "PC5", "PC6", "PC7", "PC8", "PC9", "PC10")]

    # Read ROH data
    data <- fread(roh_file)
    
    # Calculate FROH per individual
    data$FROH <- round(data$KB / 2886428, digits = 5) # Length of human autosome (GRCh38)
    data$case <- data$PHE - 1

    # Filter out invalid phenotypes
    dat <- subset(data, case != -10)

    # Merge with COVAR data
    merged_data <- merge(dat, COVAR, by = "IID")

    # Impute missing age values with the mean
    merged_impute_age <- merged_data
    merged_impute_age[is.na(merged_data$age), ]$age <- round(mean(merged_data$age, na.rm = TRUE), 0)
    
    # Generate late onset groups
    late <- subset(merged_impute_age, case == 0 | age_of_onset >= 45)
      
    # Run logistic regression (GLM) for KBAVG and other covariates
    model <- glm(case ~ KBAVG + PC1 + PC2 + PC3 + PC4 + PC5 + sex_for_qc + age, data = late, family = "binomial")

    # Extract the p-value for KBAVG
    kbavg_pvalue <- summary(model)$coefficients["KBAVG", "Pr(>|z|)"]

    # Print ancestry, ROH range, and p-value
    cat(sprintf("Ancestry: %s, Onset: Late, ROH Range: %s, KBAVG p-value: %f\n", ancestry, roh_range, kbavg_pvalue))
  }
}
```

#### For mean and SE

```python
%%R
# Set working directory
setwd("/home/jupyter/ROH_plots/")

# Define ancestry groups
ancestries <- c("AAC", "AFR", "AJ", "AMR", "CAS", "EAS", "EUR", "MDE", "SAS")

# Define file sizes (2kb to 10kb)
file_sizes <- 2:10

# Loop through ancestries
for (ancestry in ancestries) {
  cat(sprintf("\nProcessing ancestry: %s\n", ancestry))
  
  # Loop through file sizes (2KB to 10KB)
  for (size in file_sizes) {
    cat(sprintf("Processing file size: %d KB\n", size))
    
    # File paths (adjusted to working directory)
    roh_file <- sprintf("%s_ROH_%dkb.hom.indiv", ancestry, size)
    clinical_file <- "master_key_release6_final.csv"
               
    # Read data
    data <- read.table(roh_file, header = TRUE)
    clinical <- fread(clinical_file)
    colnames(clinical)[2] <- "IID"  
             
    # Calculate FROH per individual
    data$FROH <- round(data$KB / 2886428, digits = 5)  # Length of human autosome (GRCh38)
    
    # Create case variable
    data$case <- data$PHE - 1
    
    # Filter individuals with valid phenotype
    dat <- subset(data, case != -10)
    cat(sprintf("Number of phenotypes: %d\n", length(dat$case)))
    
    # Merge ROH with clinical data
    merged_data <- merge(dat, clinical, by = "IID")  
      
    # Generate early onset groups
    late <- subset(merged_data, case == 0 | age_of_onset >= 45)  
    
    # Create a tibble and calculate KBAVG mean and standard deviation
    data_tidy <- late %>%
      group_by(case) %>%
      summarise(
        KBAVG_mean = mean(KBAVG, na.rm = TRUE),
        KBAVG_se = se(KBAVG, na.rm = TRUE)
      )
    
    # Print data summary
    print(data_tidy)
    
    # Extract mean and standard deviation for cases and controls
    cases_mean <- round(data_tidy[2, 2], 1)
    cases_se <- round(data_tidy[2, 3], 1)
    controls_mean <- round(data_tidy[1, 2], 1)
    controls_se <- round(data_tidy[1, 3], 1)
    
    # Print summary message for average length of segments
    print(sprintf(
      "Average length of segments (KBAVG/AVROH) for %s %dkb: %s ± %s for Cases and %s ± %s for Controls",
      ancestry, size, cases_mean, cases_se, controls_mean, controls_se
    ))
  }
}
```
